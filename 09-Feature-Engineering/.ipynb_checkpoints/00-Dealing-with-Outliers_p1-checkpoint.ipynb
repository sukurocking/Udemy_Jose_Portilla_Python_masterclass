### Finding out the most correlated features with the destination label

```python
df['SalePrice'].info()
```

SalePrice is our label variable

```python
correlation_features = abs(df.corr()['SalePrice']).sort_values()
```

```python
correlation_features
```

As we can see, a few variables like *Gr Liv Area* and *Overall Qual* are highly correlated with *SalePrice*
Now, we need to identify outliers in the those columns specifically, since **the outliers in these highly correlated features** will mostly screw up our model

```python
#sns.scatterplot(df['SalePrice'], df['Gr Liv Area'])
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='Gr Liv Area', y='SalePrice')
```

As you can see in the above plot, we can see the 3 points, where the *Gr Liv Area* is quite high but the *SalePrice* is quite low

```python
sns.scatterplot(data=df, x='Overall Qual', y='SalePrice')
```

In our assumption, 3 outlier rows for *Overall Qual* are the same as the variable *Gr Liv Area*

```python
df[(df['SalePrice'] < 300000) & (df['Gr Liv Area'] > 4000)]
```

```python
df[(df['SalePrice'] < 200000) & (df['Overall Qual'] > 8)]
```

Grabbing the indices of the outlier variables

```python
outlier_indices = df[(df['SalePrice'] < 300000) & (df['Gr Liv Area'] > 4000)].index
```

Finally removing the outlier variables

```python
df = df.drop(outlier_indices, axis=0).shape
```
