```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../08-Linear-Regression-Models/Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop("sales", axis="columns")
y = df["sales"]

## Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

#Fitting scaler on the training dataset, scaling the training dataset and testing dataset accordingly
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

#### ElasticNet has 2 major hyperparameters that we can optimize
- alpha
- l1_ratio

```python
param_grid = {'alpha':[0.1,1,10,100], 'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
```

```python
#Now we will implement GridSearch
from sklearn.model_selection import GridSearchCV
```

```python
model = ElasticNet()
```

```python
gridsearchcv = GridSearchCV(model, param_grid, scoring = 'neg_mean_squared_error', verbose=2)
```

```python
gridsearchcv.fit(X_train, y_train)
```

```python
gridsearchcv.best_estimator_
```

```python
gridsearchcv.best_params_
```

```python
pd.DataFrame(gridsearchcv.cv_results_)
```

```python
#The below code will directly apply the best performing estimator/model and predict on the test dataset
y_pred = gridsearchcv.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MAE = mean_squared_error(y_test,y_pred) #This is our true reporting metric
MAE
```
