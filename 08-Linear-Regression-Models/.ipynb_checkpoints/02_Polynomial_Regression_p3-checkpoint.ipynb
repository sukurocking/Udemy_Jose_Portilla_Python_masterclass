#### Model optimization
Choosing optimal degree of polynomial

#### Steps
1. Choose a different degree of polynomial
2. Split polynomial features into train and test data
3. Fit on the train dataset
4. Save the errors (RMSE) for both train and test data
5. Plot the errors

```python
# Normally, we wouldnt need to go through 10 degrees of polynomial for 3 variables. 
# This is for demonstration purpose
rmse_train = []
rmse_test = []


for d in range(1,10):
    poly_converter = PolynomialFeatures(degree=d, include_bias=False)
    poly_features = poly_converter.fit_transform(X)
    
    #Splitting polynomial features into train and test data
    X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
    
    #Fitting on the training dataset
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    #Predict against both train and test data
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    
    #Computing train and test errors (rmse)
    train_error_rmse = np.sqrt(mean_squared_error(y_train, train_predictions))
    test_error_rmse = np.sqrt(mean_squared_error(y_test, test_predictions))
    
    #Saving train and test errors (rmse)
    rmse_train.append(train_error_rmse)
    rmse_test.append(test_error_rmse)
```

```python
rmse_train
```

```python
rmse_test
```

#### Plotting both train and test errors in a single plot for visualizing bias-variance trade-off

```python
plt.xlabel('Degree of polynomial')
plt.ylabel('RMSE')
plt.plot(range(1,10), rmse_train, color = 'blue')
plt.plot(range(1,10), rmse_test, color = 'red')
plt.show()
```

```python
plt.xlabel('Degree of polynomial')
plt.ylabel('RMSE')
plt.plot(range(1,6), rmse_train[:5], color = 'blue')
plt.plot(range(1,6), rmse_test[:5], color = 'red')
plt.show()
```

As we can see in the above plot, the error has exploded after 4th degree.

To be on the safer side, 3 would be the optimal polynomial degree of our polynomial model

```python
final_poly_converter = PolynomialFeatures(degree=3,include_bias = False)
```

```python
final_poly_features = final_poly_converter.fit_transform(X)
```

```python
final_poly_model = LinearRegression()
```

```python
final_poly_model.fit(final_poly_features, y)
```

```python
from joblib import dump, load
```

```python
dump(final_poly_model, 'final_poly_model.joblib')
```

```python
dump(final_poly_converter, 'final_poly_converter.joblib')
```

```python
loaded_model = load('final_poly_model.joblib')
```

```python
loaded_model.coef_
```

```python
final_poly_model.coef_
```

As we can see, the coeficients of our final model and the loaded model are the same
